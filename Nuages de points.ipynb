```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('house_pricing.csv')

data.head()
```

```python
data.info()
```

```python
plt.scatter(data.LotArea, data.SalePrice)
plt.title('Nuage de point')
plt.show()
```

```python
x = data.LotArea
y = data.SalePrice
axes = plt.gca()
axes.set_xlim(0,20000)
axes.set_ylim(0,500000)
plt.title('Nuage de point')
axes.set_xlabel('Sale Price')
axes.set_ylabel('Lot Area')
axes.scatter(x,y)
```
